```python

#slice(0, 8, 3)

y_train_1000 = y_train[slice(0, len(y_train), 40)]
x_train_1000 = x_train[slice(0, len(x_train), 40)]
y_test_1000 = y_test[slice(0, len(y_test), 40)]
x_test_1000 = x_test[slice(0, len(x_test), 40)]
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
# Choose either standardization or Normalization
# On this data Min Max Normalization is used because we need to fit Naive Bayes

# Choose between standardization and MinMAx normalization
#PredictorScaler=StandardScaler()
PredictorScaler=QuantileTransformer()

# Storing the fit object for later reference
PredictorScalerFit=PredictorScaler.fit(x_train+x_test)


x_train_1000 = PredictorScalerFit.transform(x_train_1000)
x_test_1000 = PredictorScalerFit.transform(x_test_1000)
```

```python
t = time()

callback = EarlyStopping(monitor='loss', patience=10)

i = Input(shape=(1000,), name='input')

# x = Dense(128, activation='relu')(i)
# x = Dense(64, activation='relu')(x)

x = Dense(32, activation='relu')(i)
x = Dropout(0.01)(x)
o = Dense(1, activation='softmax', name='output')(x)

model = Model(inputs=[i], outputs=[o])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(
    {'input': x_train}, 
    {'output': numpy.asarray(y_train)},
    epochs=20,
    validation_data = (x_test, numpy.asarray(y_test)),
    workers=4,
    batch_size=256,
    use_multiprocessing= True,
    callbacks=[callback], verbose=1
)

training_time = time() - t
print("train time: %0.3fs" % training_time)

y_pred = model.predict(x_test_1000)
```
